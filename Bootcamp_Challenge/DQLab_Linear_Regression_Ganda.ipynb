# Definisi
Analisis regresi digunakan untuk mengukur seberapa besar pengaruh antara variabel bebas dan variabel terikat. Regresi linear berganda merupakan model regresi yang melibatkan lebih dari satu variabel independen. Analisis regresi linear berganda dilakukan untuk mengetahui arah dan seberapa besar pengaruh variabel independen terhadap variabel dependen.


# Bentuk Umum

Bentuk umum persamaan regresi linear ganda adalah sebagai berikut
\begin{equation}
Y=b_0 + b_1 X_1 + b_2 X_2
\end{equation}

dengan 
ğ‘Œ = variabel independen
ğ‘‹ = variabel dependen
ğ‘ = koefisien regresi

# Tugas

Buatlah sebuah fungsi bernama regresi_linear_ganda dengan parameter dataX dan dataY atau <b>regresi_linear_ganda</b>(dataX, dataY) dengan keluaran berupa pasangan b0, b1, b2 dengan tipe data dict. 

Contoh :

    >>> dataX = {'x1' : [12, 14, 10, 16, 18, 24, 12], 'x2' : [10, 11, 14, 13, 15, 20, 8]}
    >>> dataY = [6, 7, 8, 8, 9, 10, 5]
    >>> b = regresi_linear_ganda(dataX, dataY)
    >>> print(b)
    ... {b0 : 2.098049, b1 : -0.0154570, b2 : 0.4390340}


```python
#Membuat fungsi regresi linier ganda
def regresi_linear_ganda(dataX, dataY):
    #Hitung rata-rata X1, Total kuadrat X1 dan total kuadrat x1
    X1_bar = sum(dataX["x1"])/len(dataX["x1"])
    init1 = []
    for i in dataX["x1"]:
        init1.append(i**2)
    X1_2 = sum(init1)
    x1_2 = X1_2 - (sum(dataX["x1"])**2)/len(dataX["x1"])
    
    #Hitung rata-rata X2, Total kuadrat X2 dan total kuadrat x2
    X2_bar = sum(dataX["x2"])/len(dataX["x2"])
    init2 = []
    for i in dataX["x2"]:
        init2.append(i**2)
    X2_2 = sum(init2)
    x2_2 = X2_2 - (sum(dataX["x2"])**2)/len(dataX["x2"])
    
    #Hitung rata_rata Y
    Y_bar = sum(dataY)/len(dataY)
    
    #Menghitung X1 kali Y dan x1 kali Y
    init3 = []
    for i in range(0,len(dataY)):
        init3.append(dataX["x1"][i]*dataY[i])
    X1_Y = sum(init3)
    
    x1_Y = X1_Y - (sum(dataX["x1"])*sum(dataY))/len(dataX["x1"])
    
    #Menghitung X2 kali Y dan x2 kali Y
    init4 = []
    for i in range(0,len(dataY)):
        init4.append(dataX["x2"][i]*dataY[i])
    X2_Y = sum(init4)
    
    x2_Y = X2_Y - (sum(dataX["x2"])*sum(dataY))/len(dataX["x2"])
    
    #Menghitung X1 kali X2
    init5 = []
    for i in range(0,len(dataX["x1"])):
        init5.append(dataX["x1"][i]*dataX["x2"][i])
    X1_X2 = sum(init5)
    
    x1_x2 = X1_X2 - (sum(dataX["x1"])*sum(dataX["x2"]))/len(dataX["x2"])
    
    #Mengitung b0, b1 dan b2
    b1 = (x2_2 * x1_Y - x1_x2 * x2_Y)/(x1_2 * x2_2 - (x1_x2**2))
    b2 = (x1_2 * x2_Y - x1_x2 * x1_Y)/(x1_2 * x2_2 - (x1_x2**2))
    b0 = Y_bar - b1 * X1_bar - b2 * X2_bar
    
    return {"b0":b0, "b1":b1, "b2":b2}
```

```python
dataX = {'x1' : [12, 14, 10, 16, 18, 24, 12], 'x2' : [10, 11, 14, 13, 15, 20, 8]}
dataY = [6, 7, 8, 8, 9, 10, 5]
c = regresi_linear_ganda(dataX, dataY)
print(c)
```
