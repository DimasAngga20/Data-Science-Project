# Definsi

Regresi Linear Sederhana merupakan metode untuk memperkirakan pengaruh variabel bebas (independen) terhadap variabel terikat (dependen). Dalam model ini banyaknya variabel bebas yang digunakan adalah 1. 

# Bentuk Umum
Bentuk umum Regresi Linear Sederhana adalah sebagai berikut

\begin{equation}
y = m\cdot x + b
\end{equation}

dengan :   
<ul>  
<li>洧녽 = variabel dependen</li>
<li>洧논 = variabel independen</li>
<li>洧녴 = slope / kemiringan (angka ini menyatakan perubahan nilai Y saat kenaikan satu satuan nilai X)</li>
<li>洧녪 = intersep, angka ini menyatakan nilai 洧녽 saat nilai 洧논 = 0</li>
</ul>

Perusahaan ingin mengetahui model regresi linearnya agar dapat memprediksi berapa advertisment spend yang harus dikeluarkan jika perusahaan menetapkan target revenue sebesar 洧녠

Jawab :     

Misalkan 

*   洧논 = advertisment spend
*   洧녽 = revenue

Diketahui banyak data 洧녵 = 7

<ol>
  <li>Hitung Total advertisment spend (洧논)</li>
  \begin{equation}
  \sum_{i=1}^{n}x_i = x_1 + x_2 + ... + x_n
  \end{equation}
  \begin{equation}
  \sum_{i=1}^{7}x_i = 400 + 380 + ... + 500 + 467
  \end{equation}
  \begin{equation}
  \sum_{i=1}^{7}x_i = 3327
  \end{equation}

  <li>Hitung Total pangkat dua dari advertisment spend (洧논^2)</li>
  \begin{equation}
  \sum_{i=1}^{n}(x_i)^2 = (x_1)^2 + (x_2)^2 + ... + (x_n)^2
  \end{equation}
  \begin{equation}
  \sum_{i=1}^{7}(x_i)^2 = (400)^2 + (380)^2 + ... + (500)^2 + (467)^2
  \end{equation}
  \begin{equation}
  \sum_{i=1}^{7}(x_i)^2 = 160000 + 144400 + ... + 250000 + 218089
  \end{equation}
  \begin{equation}
  \sum_{i=1}^{7}(x_i)^2 = 1648067
  \end{equation}

  <li>Hitung Total revenue (洧녽)</li>
  \begin{equation}
  \sum_{i=1}^{n}y_i = y_1 + y_2 + ... + y_n
  \end{equation}
  \begin{equation}
  \sum_{i=1}^{7}y_i = 551 + 480 + ... + 554 + 800
  \end{equation}
  \begin{equation}
  \sum_{i=1}^{7}y_i = 4105
  \end{equation}

  <li>Hitung Total hasil kali dari advertisment spend (洧논) dan revenue (洧녽)</li>
  \begin{equation}
  \sum_{i=1}^{n}x_i\cdot y_i = x_1\cdot y_1 + x_2\cdot y_2 + ... + x_n\cdot y_n
  \end{equation}
  \begin{equation}
  \sum_{i=1}^{7}x_i\cdot y_i = 400\cdot551 + 380\cdot480 + ... + 467\cdot 800
  \end{equation}
  \begin{equation}
  \sum_{i=1}^{7}x_i\cdot y_i = 1977084
  \end{equation}

  <li>Hitung slope 洧녴</li>
  \begin{equation}
  m = \frac{n\sum_{i=1}^{n}x_i\cdot y_i - \sum_{i=1}^{n}x_i \sum_{i=1}^{n}y_i }{n\sum_{i=1}^{n}(x_i)^2 - (\sum_{i=1}^{n}x_i)^2}
  \end{equation}

  \begin{equation}
  m = \frac{7\cdot 1977084 - 3327 \cdot 4105}{7 \cdot 1648067 - (3327)^2}
  \end{equation}

  \begin{equation}
  m = 0.3898
  \end{equation}

  <li>Hitung konstanta 洧녪</li>
  
  \begin{equation}
  b = \frac{\sum_{i=1}^{n}y_i - m\sum_{i=1}^{n}x_i}{n}
  \end{equation}

  \begin{equation}
  b = \frac{4105 - 0.3898\cdot 3327}{7}
  \end{equation}

  \begin{equation}
  b = 401.1562
  \end{equation}

  <li>Buat Model Regresinya</li>
  \begin{equation}
  y = m\cdot x + b
  \end{equation}

  \begin{equation}
  y = 0.3898\cdot x + 401.1562
  \end{equation}

  <li>Interpretasi Model</li>
  <ol>
     <li>Nilai intersep/konstanta sebesar 401.1562 artinya saat perusahaan tidak membelanjakan uangnya untuk kebutuhan iklan atau <i>advertisment spend</i> = 0, maka rata-rata perusahaan masih akan mendapatkan revenue sekitar 401.1562 satuan mata uang</li>
     <li>Koefisien slope sebesar 0.3898 artinya setiap kenaikan uang yang dibelanjakan untuk kebutuhan iklan sebesar 1 satuan, maka rata-rata revenue yang diperoleh perusahaan akan meningkat sekitar 0.3898 poin</li>
  </ol>
</ol>
